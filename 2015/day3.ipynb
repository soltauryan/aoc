```python
import aocd

data = aocd.get_data(day=3, year=2015)
data
```

```python
def move(char):
    if char == '>':
        return 1, 0
    elif char == '<':
        return -1, 0
    elif char == '^':
        return 0, 1
    elif char == 'v':
        return 0, -1
    else:
        print(char)
        raise ValueError

def main(data):
    coord_set = set()
    x_coord = 0
    y_coord = 0

    for char in data:
        x_delt, y_delt = move(char)
        x_coord += x_delt
        y_coord += y_delt
        coord_set.add((x_coord, y_coord))
    
    print(sorted(coord_set))

    return len(coord_set) + 1

main(data)
```

```python
data_test = '^v^v^v^v^v'

main(data_test)
```

```python
def split_instructions(data):
    ins_one = ''
    ins_two = ''

    for i, char in enumerate(data):
        if i % 2 == 0:
            ins_one += char
        else:
            ins_two += char

    return ins_one, ins_two

split_instructions(data_test)
```

```python
def main(data):
    coord_set = set()
    coord_set.add((0, 0))
    x_coord = 0
    y_coord = 0

    real_santa, robo_santa = split_instructions(data)

    for char in real_santa:
        x_delt, y_delt = move(char)
        x_coord += x_delt
        y_coord += y_delt
        coord_set.add((x_coord, y_coord))
    
    x_coord = 0
    y_coord = 0

    for char in robo_santa:
        x_delt, y_delt = move(char)
        x_coord += x_delt
        y_coord += y_delt
        coord_set.add((x_coord, y_coord))

    print(sorted(coord_set))

    return len(coord_set)

main(data)
```
