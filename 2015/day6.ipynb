```python
import aocd
import re
```

```python

data = aocd.get_data(day=6, year=2015)
data = data.splitlines()
```

```python
data
```

```python
s = data[7]

def get_instructions(s):
    pattern = r"^(\D+)(\d+,\d+) through (\d+,\d+)"
    match = re.search(pattern, s)

    if match:
        instruction = match.group(1).strip()
        start_coord = match.group(2)
        end_coord = match.group(3)

    return instruction, start_coord, end_coord

get_instructions(s)
```

```python
class Light():
    def __init__(self):
        self.on = False

    def turn_on(self):
        self.on = True

    def turn_off(self):
        self.on = False

    def toggle(self):
        if self.on:
            self.on = False
        else:
            self.on = True


class Grid():
    def __init__(self, x_lights=1000, y_lights=1000):
        self.light_grid = [[Light() for _ in range(x_lights)] for _ in range(y_lights)]
    
    def lights_lit(self):
        total = 0
        for light_row in self.light_grid:
            # print(light_row[0].on)
            total += sum([l.on for l in light_row])
        return total
    

g = Grid()
g.lights_lit()
```

```python
def main(data):
    g = Grid()
    
    for instruct in data:
        instruction, start_coord, end_coord = get_instructions(instruct)
        start_row, start_col = start_coord.split(",")
        start_row, start_col = int(start_row), int(start_col)
        end_row, end_col = end_coord.split(",")
        end_row, end_col = int(end_row), int(end_col)

        for row in range(start_row, end_row+1):
            for col in range(start_col, end_col+1):
                if instruction == 'turn on':
                    g.light_grid[row][col].turn_on()
                elif instruction == 'turn off':
                    g.light_grid[row][col].turn_off()
                else:
                    g.light_grid[row][col].toggle()

    return g.lights_lit()

main(data)
```

```python
class Light():
    def __init__(self):
        self.brightness = 0

    def turn_on(self):
        self.brightness += 1

    def turn_off(self):
        self.brightness = max(self.brightness - 1, 0)

    def toggle(self):
        self.brightness += 2


class Grid():
    def __init__(self, x_lights=1000, y_lights=1000):
        self.light_grid = [[Light() for _ in range(x_lights)] for _ in range(y_lights)]
    
    def lights_lit(self):
        total = 0
        for light_row in self.light_grid:
            # print(light_row[0].on)
            total += sum([l.brightness for l in light_row])
        return total
    
def main(data):
    g = Grid()
    
    for instruct in data:
        instruction, start_coord, end_coord = get_instructions(instruct)
        start_row, start_col = start_coord.split(",")
        start_row, start_col = int(start_row), int(start_col)
        end_row, end_col = end_coord.split(",")
        end_row, end_col = int(end_row), int(end_col)

        for row in range(start_row, end_row+1):
            for col in range(start_col, end_col+1):
                if instruction == 'turn on':
                    g.light_grid[row][col].turn_on()
                elif instruction == 'turn off':
                    g.light_grid[row][col].turn_off()
                else:
                    g.light_grid[row][col].toggle()

    return g.lights_lit()

main(data)
```
