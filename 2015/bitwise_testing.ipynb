```python
data = aocd.get_data(day=7, year=2015).splitlines()
data
```

```python
wire_dict = {}

def bit_not(n, numbits=16):
    return (1 << numbits) - 1 - n

for instruct in data:
    instruction, var = instruct.split(" -> ")
    print(instruction, var)
    
    if instruction.isnumeric():
        wire_dict[var] = int(instruction)
    
    elif instruction.find(' AND ') != -1:
        num1, num2 = instruction.split(" AND ")
        print(num1, num2)
        if num1.isalpha():
            num1 = wire_dict[num1]
        if num2.isalpha():
            num2 = wire_dict[num2]
        wire_dict[var] = num1 & num2        

    elif instruction.find(' OR ') != -1:
        num1, num2 = instruction.split(" OR ")
        if num1.isalpha():
            num1 = wire_dict[num1]
        if num2.isalpha():
            num2 = wire_dict[num2]
        wire_dict[var] = num1 | num2        

    elif instruction.find(' LSHIFT ') != -1:
        num1, num2 = instruction.split(" LSHIFT ")
        if num1.isalpha():
            num1 = wire_dict[num1]
        else:
            num1 = int(num1)
        if num2.isalpha():
            num2 = wire_dict[num2]
        else:
            num2 = int(num2)
        wire_dict[var] = num1 << num2      

    elif instruction.find(' RSHIFT ') != -1:
        num1, num2 = instruction.split(" RSHIFT ")
        if num1.isalpha():
            num1 = wire_dict[num1]
        else:
            num1 = int(num1)
        if num2.isalpha():
            num2 = wire_dict[num2]
        else:
            num2 = int(num2)
        wire_dict[var] = num1 >> num2 

    elif instruction.find('NOT ') != -1:
        num = instruction[4:]
        if num.isalpha():
            num = wire_dict[num]
        else:
            num = int(num)
        wire_dict[var] = bit_not(num)

wire_dict
```
