```python
def main(data):
    num = 1
    while True:
        hash_str = hashlib.md5(f"{data}{str(num)}".encode()).hexdigest()
        if hash_str[0:5] == '00000':
            return f"{data}{num}"

        if num % 100_000 == 0:
            print(num)
            print(hash_str)

        if num == 2000000:
            return 0
        
        num +=1

main(data)
```

```python
def main(data):
    num = 1
    while True:
        hash_str = hashlib.md5(f"{data}{str(num)}".encode()).hexdigest()
        if hash_str[0:6] == '000000':
            return f"{data}{num}"

        if num % 1_000_000 == 0:
            print(num)
            print(hash_str)

        if num == 200000000:
            return 0
        
        num +=1

main(data)
```
