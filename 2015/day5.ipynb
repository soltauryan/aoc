```python
def get_substrings(string):
    for i in range(len(string) - 1):
        yield string[i : i+2]

gen = get_substrings("pythonxy")

for i in gen:
    print(i in ['ab', 'cd', 'pq', 'xy'])
```

```python
def main(data):
    vowels = set('aeiou')
    nice_count = 0

    for s in data:
        check_one = False
        check_two = False
        check_three = True
        
        # First rule check
        c = Counter(s)
        vowel_count = c.get('a', 0) + c.get('e', 0) + c.get('i', 0) + c.get('o', 0) + c.get('u', 0)
        if vowel_count >= 3:
            check_one = True

        gen = get_substrings(s)
        for substring in gen:
            if substring[0] == substring[1]:
                check_two = True
            if substring in ['ab', 'cd', 'pq', 'xy']:
                check_three = False

        if all([check_one, check_two, check_three]):
            nice_count += 1


    return nice_count
        

# main(['aaa'])
main(data)
```

# P2 Regex

```python
import re

pattern_pair = r"(\D{2}).*\1"
pattern_sandwich = r"(\D).\1"

def main(data):
    nice_count = 0

    for s in data:
        if re.search(pattern_pair, s) and re.search(pattern_sandwich, s):
            nice_count += 1
    
    return nice_count

main(data)
```
