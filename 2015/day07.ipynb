```python
def not_op(num):
    return ~num & 0xFFFF

def and_op(num1, num2):
    return (num1 & num2) & 0xFFFF

def or_op(num1, num2):
    return (num1 | num2) & 0xFFFF

def l_shift_op(num, shift):
    return (num << shift) & 0xFFFF

def r_shift_op(num, shift):
    return (num >> shift) & 0xFFFF
```

```python
signals = {}

def ingest(data, p2):
    wire_instructions = {}
    for line in data:
        instruction, assign = line.split(" -> ")
        if p2 and assign == 'b':
            wire_instructions['956'] = assign
        else:
            wire_instructions[instruction] = assign
    return wire_instructions


def main(data, p2=False):
    wire_dict = ingest(data, p2)
    value_dict = {}

    while len(wire_dict) != 0:
        num_list = []
        for k, v in wire_dict.items():
            # Check for digits
            if k.isdigit():
                num_list.append(k)
                value_dict[v] = k
            # Check if both sides are in value_dict
            if k.isalpha() and k in value_dict:
                num_list.append(k)
                value_dict[v] = value_dict[k]


            # Check for NOT operator
            if "NOT" in k:
                _, n_v = k.split("NOT ")
                if n_v in value_dict:
                    num_list.append(k)
                    value_dict[v] = not_op(int(value_dict[n_v]))


            # Check for AND operator
            if " AND " in k:
                a, b = k.split(" AND ")
                if a in value_dict and b in value_dict:
                    num_list.append(k)
                    value_dict[v] = and_op(int(value_dict[a]), int(value_dict[b]))
                elif a in value_dict and b.isdigit():
                    num_list.append(k)
                    value_dict[v] = and_op(int(value_dict[a]), int(b))
                elif b in value_dict and a.isdigit():
                    num_list.append(k)
                    value_dict[v] = and_op(int(a), int(value_dict[b]))

            
            # Check for OR operator
            if " OR " in k:
                a, b = k.split(" OR ")
                if a in value_dict and b in value_dict:
                    num_list.append(k)
                    value_dict[v] = or_op(int(value_dict[a]), int(value_dict[b]))
                elif a in value_dict and b.isdigit():
                    num_list.append(k)
                    value_dict[v] = or_op(int(value_dict[a]), int(b))
                elif b in value_dict and a.isdigit():
                    num_list.append(k)
                    value_dict[v] = or_op(int(a), int(value_dict[b]))

            # Check for LSHIFT operator
            if " LSHIFT " in k:
                a, b = k.split(" LSHIFT ")
                if a in value_dict:
                    num_list.append(k)
                    value_dict[v] = l_shift_op(int(value_dict[a]), int(b))

            # Check for RSHIFT operator
            if " RSHIFT " in k:
                a, b = k.split(" RSHIFT ")
                if a in value_dict:
                    num_list.append(k)
                    value_dict[v] = r_shift_op(int(value_dict[a]), int(b))


        for n in num_list:
            wire_dict.pop(n)



    return value_dict['a']

main(data, p2=True)
```
