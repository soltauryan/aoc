```python
import aocd

data = aocd.get_data(day=2, year=2015)
data = data.splitlines()
```

```python

def calc_sqft(line):
    a, b, c = line.split('x')
    a, b, c = int(a), int(b), int(c)
    ab_area = 2 * a * b
    ac_area = 2 * a * c
    bc_area = 2 * b * c
    min_area = min([ab_area, ac_area, bc_area]) / 2
    return ab_area + ac_area + bc_area + min_area
```

```python
s1, s2 = '2x3x4', '1x1x10'

calc_sqft(s1), calc_sqft(s2)
```

```python
def main(data):
    total = 0
    for box in data:
        total += calc_sqft(box)
    return total

main(data)
```

```python
def calc_ribbon_ft(line):
    a, b, c = line.split('x')
    a, b, c = int(a), int(b), int(c)
    min_list = sorted([a, b, c])
    min_one, min_two = min_list[0], min_list[1]
    return (2 * min_one) + (2 * min_two) + (a * b * c)
```

```python
def main(data):
    total = 0
    for box in data:
        total += calc_ribbon_ft(box)
    return total

main(data)
```
