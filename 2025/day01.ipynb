```python
def main(data):
    dial_location = 50
    zero_count = 0
    dial_track = []

    for move in data:
        dir = move[0]
        clicks = int(move[1:])

        # 1. Handle "bulk" rotations
        # Every 100 clicks guarantees we pass 0 exactly once
        zero_count += clicks // 100

        # 2. Simulate the remaining clicks one by one
        remaining_clicks = clicks % 100
        
        for _ in range(remaining_clicks):
            if dir == 'L':
                dial_location = (dial_location - 1) % 100
            else: # 'R'
                dial_location = (dial_location + 1) % 100
            
            # Count if we land on 0 during these specific clicks
            if dial_location == 0:
                zero_count += 1

        dial_track.append(dial_location)

    return dial_location, zero_count, dial_track

main(data)
```
