```python
def create_grid(max):
    return [["." for _ in range(max)] for _ in range(max)]

def print_grid(grid):
    for row in grid:
        print(row)

def fill_insides_scanline(grid):
    rows = len(grid)
    cols = len(grid[0])
    
    for r in range(rows):
        inside = False
        for c in range(cols):
            cell = grid[r][c]
            
            # Logic: If we hit a vertical boundary, toggle state.
            # Warning: This simplistic check assumes any '#' is a vertical wall.
            # In complex shapes, "######" is a horizontal wall and shouldn't toggle.
            # You usually need to know the source geometry to do this perfectly on a grid.
            if cell == "#":
                # You might need to check if this '#' connects North/South 
                # to determine if it's a vertical boundary crossing.
                inside = not inside 
            
            elif inside and cell == ".":
                grid[r][c] = "X"
                
    return grid

from collections import deque

def fill_insides_flood(grid):
    rows = len(grid)
    cols = len(grid[0])
    
    # 1. Start a BFS/Flood fill from the top-left corner (0,0)
    # Note: This assumes the shape doesn't touch the edges of the grid.
    # If it does, you need to increase 'max' by 2 and offset coords by 1.
    queue = deque([(0, 0)])
    visited = set()
    
    while queue:
        r, c = queue.popleft()
        
        if (r, c) in visited:
            continue
        visited.add((r, c))
        
        # If we hit a wall (#) or your existing outline (X), stop this path
        if grid[r][c] != ".":
            continue
            
        # Mark as 'O' for Outside (temporary)
        grid[r][c] = "O"
        
        # Check neighbors
        for dr, dc in [(0, 1), (0, -1), (1, 0), (-1, 0)]:
            nr, nc = r + dr, c + dc
            if 0 <= nr < rows and 0 <= nc < cols:
                queue.append((nr, nc))

    # 2. Iterate through grid. Anything that is still '.' is Inside.
    for r in range(rows):
        for c in range(cols):
            if grid[r][c] == ".":
                grid[r][c] = "X" # This is the inside!
            elif grid[r][c] == "O":
                grid[r][c] = "." # Restore outside to empty
                
    return grid

def p2(data):
    # max = 14
    max = 100_000
    coords = ingest(data)
    coords.append(coords[0])
    n = len(coords)

    grid = create_grid(max)

    for i in range(0, n-1):
        j = i + 1
        c1, c2 = coords[i], coords[j]
        c1_col, c1_row = c1[0], c1[1]
        c2_col, c2_row = c2[0], c2[1]

        grid[c1_row][c1_col] = "#"
        grid[c2_row][c2_col] = "#"

        if c1_row == c2_row:
            if c2_col < c1_col:
                for j in range(c1_col-1, c2_col, -1):
                    grid[c1_row][j] = "X"
            else:
                for j in range(c1_col+1, c2_col):
                    grid[c1_row][j] = "X"
        else:
            if c2_row < c1_row:
                for j in range(c1_row-1, c2_row, -1):
                    grid[j][c1_col] = "X"
            else:
                for j in range(c1_row+1, c2_row):
                    grid[j][c1_col] = "X"

    grid = fill_insides_flood(grid)
            
    print_grid(grid)      



    return 0

p2(data)
```
