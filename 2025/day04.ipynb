```python
test = """..@@.@@@@.
@@@.@.@.@@
@@@@@.@.@@
@.@@@@..@.
@@.@@@@.@@
.@@@@@@@.@
.@.@.@.@@@
@.@@@.@@@@
.@@@@@@@@.
@.@.@@@.@.""".splitlines()

test2 = """..........
..........
..@@@.....
..@@@@....
...@@@@...
...@@@@@..
...@.@.@@.
...@@.@@@.
...@@@@@@.
....@@@...""".splitlines()



test = [list(row) for row in test]
data = [list(row) for row in data]
test2 = [list(row) for row in test2]

def return_surrounding_coord(grid, row, col):
    grid_row_max_len = len(grid) - 1
    grid_col_max_len = len(grid[0]) - 1
    coord_set = set()
    for row_i in range(row-1, row+2):
        for col_i in range(col-1, col+2):
            if row_i < 0 or col_i < 0 or row_i > grid_row_max_len or col_i > grid_col_max_len:
                continue
            elif row_i == row and col_i == col:
                continue
            else:
                coord_set.add((row_i, col_i))


    return coord_set

def main(grid):
    tp_set = set()
    tp_to_remove_set = set()
    tp_available_count = 0

    for row in range(len(grid)):
        for col in range(len(grid[0])):
            if grid[row][col] == "@":
                tp_set.add((row, col))

    for tp in tp_set:
        surrouding_coord = return_surrounding_coord(grid, tp[0], tp[1])
        if len(surrouding_coord) < 4:
            tp_available_count += 1
            tp_to_remove_set.add((tp[0],tp[1]))
            continue

        tp_surround_count = 0
        for coord in surrouding_coord:
            if grid[coord[0]][coord[1]] == '@':
                tp_surround_count += 1
        
        if tp_surround_count < 4:
            tp_available_count += 1
            tp_to_remove_set.add((tp[0], tp[1]))

    for tp in tp_to_remove_set:
        grid[tp[0]][tp[1]] = '.'
    
    return tp_available_count, grid

loop_bool = True
acc = 0
count, grid = main(data)
acc += count
i = 0

while loop_bool:
    count, grid = main(grid)
    acc += count
    if count == 0:
        loop_bool = False
    i += 1


i, acc
```
