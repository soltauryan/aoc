```python
def ingest_data(raw_data):
    fresh, ids = raw_data.split("\n\n")
    # fresh = list(fresh)
    fresh = fresh.splitlines()
    ids = ids.splitlines()
    return fresh, ids

def main(data):
    fresh, ids = ingest_data(data)

    # Create set of fresh fruits
    fresh_set = set()
    for r in fresh:
        start, stop = r.split("-")
        start, stop = int(start), int(stop)
        for i in range(start, stop+1):
            fresh_set.add(i)
    
    # Count ids in fresh set
    fresh_id_count = 0
    for id in ids:
        if int(id) in fresh_set:
            fresh_id_count += 1

    return fresh_id_count

main(test)
```
