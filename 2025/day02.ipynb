```python
s = data[5].split('-')
s = "95-115"
s = s.split("-")
num1, num2 = s[0], s[1]
print(num1, num2, '\n')


def main(data):
    invalid_list = []
    for line in data:
        s = line.split("-")
        num1, num2 = s[0], s[1]
        for num in range(int(num1), int(num2) + 1):
            str_num = str(num)
            if len(str_num) % 2 != 0:
                continue
            else:
                split = int(len(str_num) / 2)
                if str_num[:split] == str_num[split:]:
                    invalid_list.append(num)
    return sum(invalid_list)

main(data)
```
