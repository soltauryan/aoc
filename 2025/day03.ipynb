```python
import aocd

lines = aocd.get_data(day=3, year=2025)
data = lines.splitlines()
len(data)
```

```python
test = """987654321111111
811111111111119
234234234234278
818181911112111""".splitlines()

# test = """987654321111111""".splitlines()
total = 0
for line in data:
    max_num = 0
    for i in range(len(line)):
        for j in range(i, len(line)):
            if i == j:
                continue
            current_num = int(line[i]+line[j])
            max_num = max(max_num, current_num)
        
    total += max_num

total
```

# P2

```python
jolt_sum = 0
for line in data:
    t = list(line)
    output_jolt = ""

    i=11
    while i >=0:
        t_list = list(t[:len(t)-i])
        output_jolt += str(max(t_list))
        t_max_i = t.index(max(t_list))
        for j in range(0, t_max_i+1):
            t.pop(0)
        i -= 1
    jolt_sum += int(output_jolt)

jolt_sum
```
