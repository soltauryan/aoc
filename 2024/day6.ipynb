```python
import multiprocessing
from functools import partial

# --- Your Original Guard Class (Unchanged) ---
class Guard:
    def __init__(self, cur_i, cur_j, direction, list_grid) -> None:
        self.cur_i = cur_i
        self.cur_j = cur_j
        self.direction = direction
        self.list_grid = list_grid
        self.in_grid = True
        self.move_dict = {
            '^':(-1, 0), '>':(0, 1), 'v':(1, 0), '<':(0, -1),
        }
        # Remove guard from list_grid once coordinates are determined
        self.list_grid[cur_i][cur_j] = "."
        self.i_limit = len(list_grid) - 1
        self.j_limit = len(list_grid[0]) - 1

    def next_coordinate(self):
        next_i = self.cur_i + self.move_dict[self.direction][0]
        next_j = self.cur_j + self.move_dict[self.direction][1]
        return next_i, next_j

    def check_if_next_move_outside_grid(self):
        test_i, test_j = self.next_coordinate() 
        if test_i < 0 or test_i > self.i_limit:
            self.in_grid = False
            return True
        elif test_j < 0 or test_j > self.j_limit:
            self.in_grid = False
            return True
        return False

    def return_char_in_front_of_guard(self):
        next_i, next_j = self.next_coordinate()
        return self.list_grid[next_i][next_j]

    def turn_right(self):
        turn_dict = {'^':'>', '>':'v', 'v':'<', '<':'^'}
        self.direction = turn_dict[self.direction]

    def move(self):
        next_i, next_j = self.next_coordinate()
        self.cur_i = next_i
        self.cur_j = next_j

    def step(self):
        next_char = self.return_char_in_front_of_guard()
        if next_char == '.':
            self.move()
        elif next_char == '#':
            self.turn_right()
        self.check_if_next_move_outside_grid()

# --- Helper Functions ---

def ingest_input_str(input_str):
    list_grid = input_str.split("\n")
    return [list(l) for l in list_grid]

def find_guard(list_grid):
    for i, sublist in enumerate(list_grid):
        for j, char in enumerate(sublist):
            if char in ['^', '>', 'v', '<']:
                return Guard(i, j, char, list_grid)
    raise ValueError("Input string doesn't contain guard")

# --- THE WORKER FUNCTION ---
# This function runs on a separate processor core.
# It needs everything required to run the simulation in isolation.
def check_obstacle_at_position(input_str, target_i, target_j):
    # 1. Create a fresh grid for this process
    temp_list_grid = ingest_input_str(input_str)
    
    # 2. Place the obstacle
    temp_list_grid[target_i][target_j] = "#"
    
    # 3. Initialize guard
    # Note: find_guard is slightly expensive to run every time. 
    # In a hyper-optimized version, we would pass the guard's start pos 
    # to this function to avoid searching the grid every time.
    temp_guard = find_guard(temp_list_grid)
    
    # 4. Run Simulation
    step = 0
    step_limit = 500000 # Reduced for demonstration, original was 1_000_000
    
    # OPTIMIZATION: 
    # Using a set to track visited states (pos + direction) is 
    # significantly faster than running a step counter to 1,000,000.
    visited_states = set()
    
    while step < step_limit:
        if not temp_guard.in_grid:
            return 0 # False (Did not loop)
        
        # State tracking for loop detection
        state = (temp_guard.cur_i, temp_guard.cur_j, temp_guard.direction)
        if state in visited_states:
            return 1 # True (Loop detected)
        visited_states.add(state)
        
        temp_guard.step()
        step += 1
        
    # If we hit step limit, we assume loop (or just return 0 if unsure)
    return 1 if step == step_limit else 0

# --- Main Execution ---

def main_parallel(input_str):
    list_grid = ingest_input_str(input_str)
    
    # 1. Find initial guard position to avoid placing obstacle on the guard
    initial_guard = find_guard(ingest_input_str(input_str))
    g_start_i, g_start_j = initial_guard.cur_i, initial_guard.cur_j
    
    # 2. Build a list of tasks (coordinates to test)
    tasks = []
    rows = len(list_grid)
    cols = len(list_grid[0])
    
    for i in range(rows):
        for j in range(cols):
            # Skip guard start position and existing obstacles
            if (i == g_start_i and j == g_start_j) or list_grid[i][j] == '#':
                continue
            
            # Arguments tuple: (input_str, i, j)
            tasks.append((input_str, i, j))

    print(f"Processing {len(tasks)} possibilities on {multiprocessing.cpu_count()} cores...")

    # 3. Initialize the Pool
    # We use 'starmap' because our worker function takes multiple arguments
    with multiprocessing.Pool() as pool:
        results = pool.starmap(check_obstacle_at_position, tasks)

    # 4. Sum the results (1s and 0s)
    total_loops = sum(results)
    return total_loops

# Crucial: Multiprocessing requires this check on Windows/Mac
if __name__ == "__main__":
    # Example tiny input
    sample_input = """....#.....
.........#
..........
..#.......
.......#..
..........
.#..^.....
........#.
#.........
......#..."""
    
    result = main_parallel(input_str)
    print(f"Found {result} looping positions.")
```
