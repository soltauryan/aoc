```python
report_list = input_str.split("\n")
report_list
```

```python
test = [item.split(' ') for item in report_list]
report_list_int = []
for list in test:
    report_list_int.append([int(item) for item in list])

len(report_list_int)
```

```python
safe_list = []

def greater_than_three_check(val_1, val_2):
    if abs(val_1 - val_2) > 3:
        return False
    else:
        return True


for sublist in report_list_int:
    last_index = len(sublist) - 2
    increasing = None
    for i in range(len(sublist) - 1):
        val_1, val_2 = sublist[i], sublist[i + 1]
        # First let establish if increasing or decreasing
        if increasing is None:
            if val_1 > val_2:
                increasing = False
            elif val_1 < val_2:
                increasing = True
            else:
                safe_list.append(False)
                break

        # Handle increasing case
        if increasing:
            if val_1 < val_2 and greater_than_three_check(val_1, val_2):
                pass
            else:
                safe_list.append(False)
                break


        # Handle decreasing case
        if not increasing:
            if val_1 > val_2 and greater_than_three_check(val_1, val_2):
                pass
            else:
                safe_list.append(False)
                break
        
        if i == last_index:
            safe_list.append(True)
            break
```

```python
def greater_than_three_check(val_1, val_2):
    if abs(val_1 - val_2) > 3:
        return False
    else:
        return True

def check_lists(sublist):

    last_index = len(sublist) - 2
    increasing = None
    for i in range(len(sublist) - 1):
        val_1, val_2 = sublist[i], sublist[i + 1]
        # First let establish if increasing or decreasing
        if increasing is None:
            if val_1 > val_2:
                increasing = False
            elif val_1 < val_2:
                increasing = True
            else:
                return False

        # Handle increasing case
        if increasing:
            if val_1 < val_2 and greater_than_three_check(val_1, val_2):
                pass
            else:
                return False


        # Handle decreasing case
        if not increasing:
            if val_1 > val_2 and greater_than_three_check(val_1, val_2):
                pass
            else:
                return False
        
        if i == last_index:
            return True


safe_list = []

for i in range(len(report_list_int)):
    report = report_list_int[i]
    list_len = len(report)
    report_safe_list = []

    # Check the list as is
    report_safe_list.append(check_lists(report))

    # Check every variation of the list minus 1 item
    for i in range(len(report)):
        subreport = report[:i] + report[i + 1:]
        report_safe_list.append(check_lists(subreport))
    
    if any(report_safe_list):
        safe_list.append(True)
    else:
        safe_list.append(False)

print(safe_list)
```
