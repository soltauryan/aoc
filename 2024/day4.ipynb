```python
def input_to_puzzle_list(input_puzzle):
    test_puzzle = input_puzzle.split("\n")
    test_puzzle = [list(p) for p in test_puzzle]
    
    return test_puzzle

def return_xy_puzzle_dimensions(puzzle):
    return len(puzzle[0]), len(puzzle)

def return_coord_for_first_letter(puzzle, letter):
    x, y = return_xy_puzzle_dimensions(puzzle)
    return_list = []

    for i in range(x):
        for j in range(y):         
            if puzzle[j][i] == letter:
                return_list.append((i, j))

    return return_list

def check_direction(puzzle, word, x, y, direction):
    dir_dict = {
        "r": (1, 0),
        "l": (-1, 0),
        "u": (0, -1),
        "d": (0, 1),
        "ul": (-1, -1),
        "ur": (1, -1),
        "dl": (-1, 1),
        "dr": (1, 1),
    }
    direction_tuple = dir_dict.get(direction)
    if direction_tuple == -1:
        raise AttributeError("Direction needs to be ul, ur, dl, dr")
    x_d = direction_tuple[0]
    y_d = direction_tuple[1]    

    word_len = len(word)
    puzzle_x_len, puzzle_y_len = return_xy_puzzle_dimensions(puzzle)

    # Can the word fit?
    if x + (word_len - 1) * x_d >= puzzle_x_len or x + (word_len - 1) * x_d < 0:
        return False
    if y + (word_len - 1) * y_d >= puzzle_y_len or y + (word_len - 1) * y_d < 0:
        return False
    # Since it fits, check that following letters are correct
    if puzzle[y+1*y_d][x+1*x_d] != 'M':
        return False
    if puzzle[y+2*y_d][x+2*x_d] != 'A':
        return False
    if puzzle[y+3*y_d][x+3*x_d] != 'S':
        return False
    
    # Return True if word is there
    return True
```

```python
def main(puzzle_input):
    dir_list = ["r", "l", "u", "d", "ul", "ur", "dl", "dr"]
    word_match_count = 0

    puzzle = input_to_puzzle_list(puzzle_input)
    coord_list = return_coord_for_first_letter(puzzle, "X")
    for coord in coord_list:
        for dir in dir_list:
            try:
                if check_direction(puzzle, "XMAS", coord[0], coord[1], dir):
                    word_match_count += 1
            except:
                print(coord, dir)

    return word_match_count

main(test_puzzle_input)
```

```python
def input_to_puzzle_list(input_puzzle):
    test_puzzle = input_puzzle.split("\n")
    test_puzzle = [list(p) for p in test_puzzle]
    
    return test_puzzle

def return_xy_puzzle_dimensions(puzzle):
    return len(puzzle[0]), len(puzzle)

def return_coord_for_first_letter(puzzle, letter):
    x, y = return_xy_puzzle_dimensions(puzzle)
    return_list = []

    for i in range(x):
        for j in range(y):         
            if puzzle[j][i] == letter:
                return_list.append((i, j))

    return return_list

def check_direction(puzzle, word, x, y, direction):
    dir_dict = {
        "r": (1, 0),
        "l": (-1, 0),
        "u": (0, -1),
        "d": (0, 1),
        "ul": (-1, -1),
        "ur": (1, -1),
        "dl": (-1, 1),
        "dr": (1, 1),
    }
    direction_tuple = dir_dict.get(direction)
    if direction_tuple == -1:
        raise AttributeError("Direction needs to be ul, ur, dl, dr")
    x_d = direction_tuple[0]
    y_d = direction_tuple[1]    

    word_len = len(word)
    puzzle_x_len, puzzle_y_len = return_xy_puzzle_dimensions(puzzle)

    # Can the word fit?
    if x + (word_len - 1) * x_d >= puzzle_x_len or x + (word_len - 1) * x_d < 0:
        return False, (-1, -1)
    if y + (word_len - 1) * y_d >= puzzle_y_len or y + (word_len - 1) * y_d < 0:
        return False, (-1, -1)
    # Since it fits, check that following letters are correct
    if puzzle[y+1*y_d][x+1*x_d] != 'A':
        return False, (-1, -1)
    if puzzle[y+2*y_d][x+2*x_d] != 'S':
        return False, (-1, -1)
    
    # Return True and 'A' coordinates if word is there
    return True, (x+1*x_d, y+1*y_d)
```

```python
def main(puzzle_input):
    dir_list = ["ul", "ur", "dl", "dr"]
    mas_a_letter_coord_dict = {}

    puzzle = input_to_puzzle_list(puzzle_input)
    coord_list = return_coord_for_first_letter(puzzle, "M")
    for coord in coord_list:
        for dir in dir_list:
            try:
                is_there, a_coord = check_direction(puzzle, "MAS", coord[0], coord[1], dir)
                if is_there:
                    mas_a_letter_coord_dict[a_coord] = mas_a_letter_coord_dict.get(a_coord, 0)+1
            except:
                print(coord, dir)

    # Filter dict for coords with > 1 occurence, signifiying an X
    filtered_dict = dict((k, v) for k, v in mas_a_letter_coord_dict.items() if v > 1)

    return len(filtered_dict)

main(puzzle_input)
```
