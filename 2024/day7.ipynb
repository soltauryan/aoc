# Gemeni multiprocessing

```python
import multiprocessing

# 1. WORKER FUNCTION
# This function handles the logic for a single equation.
# It uses recursion with pruning (stopping early if value > target)
# which is significantly faster than checking every combination.
def solve_recursive(target, current_val, numbers, index):
    # Base case: we've used all numbers
    if index == len(numbers):
        return current_val == target
    
    # Pruning: If we are already above the target, stop (since nums are positive)
    if current_val > target:
        return False
        
    next_num = numbers[index]
    
    # Try Addition (+)
    if solve_recursive(target, current_val + next_num, numbers, index + 1):
        return True
        
    # Try Multiplication (*)
    if solve_recursive(target, current_val * next_num, numbers, index + 1):
        return True
        
    # Try Concatenation (|)
    # Math trick: 12 || 34 -> 12 * 100 + 34
    # This is faster than int(str(val) + str(next_num))
    digits = 1
    temp = next_num
    while temp >= 10:
        temp //= 10
        digits += 1
        
    concatenated = current_val * (10**digits) + next_num
    if solve_recursive(target, concatenated, numbers, index + 1):
        return True

    return False

# 2. WRAPPER
# Parses one line and calls the solver.
def process_line(line):
    if not line.strip():
        return 0
        
    parts = line.split(": ")
    target = int(parts[0])
    nums = [int(n) for n in parts[1].split()]
    
    # Start recursion: current_val is the first number, index starts at 1
    if solve_recursive(target, nums[0], nums, 1):
        return target
    return 0

# 3. MAIN ENTRY POINT
def main():

    lines = puzzle_input.strip().split("\n")
    
    # Use all available CPU cores
    # On Windows, this block is critical to avoid recursive spawning
    with multiprocessing.Pool() as pool:
        # map distributes the list 'lines' across processes
        results = pool.map(process_line, lines)

    print(f"Total calibration result: {sum(results)}")

if __name__ == '__main__':
    multiprocessing.freeze_support() # Optional, good practice for Windows executables
    main()
```
